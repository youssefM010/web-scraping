```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time

while True:
    url = 'https://www.worldometers.info/coronavirus/country/egypt/'
    response = requests.request('GET', url)

    soup = BeautifulSoup(response.text, 'html.parser')

    data = soup.find_all('div', attrs={'class':'col-md-8'})

    data_a = data[0].find('div').text

    nu = data_a.split()

    cases = nu[29]
    recovered = nu[33]

    print(cases)
    print(recovered)
    
    f = open('cases.txt', 'a')
    now = datetime.now()
    time_now = now.strftime("%d/%m/%Y")
    output = 'Total cases= '+cases+','+'Recovered= '+recovered+', at '+time_now+'\n\n'
    f.write(output)
    f.close()
    time.sleep(86400)
```
